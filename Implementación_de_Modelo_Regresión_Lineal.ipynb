## Regresión Lineal Simple

Específicamente, apliqué un modelo de regresión lineal para investigar la relación entre la estatura y el peso, utilizando técnicas de aprendizaje automático. Este modelo nos proporciona una herramienta para comprender cómo una variable, en este caso la estatura, puede prever o influir en otra variable, que en este contexto es el peso. Este enfoque no solo nos permite aplicar conceptos teóricos aprendidos en clase, sino que también nos brinda la oportunidad de explorar y analizar prácticamente la interrelación entre estas variables en el contexto del grupo estudiado.

## Datos

Datos.png

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt

# Datos
estatura = np.array([1.86, 1.60, 1.57, 1.80, 1.67, 1.55, 1.62, 1.70, 1.55, 1.57]).reshape(-1, 1)
peso = np.array([90, 60, 58, 85, 60, 56, 65, 70, 55, 60])

# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(estatura, peso)

# Hacer predicciones
predicciones_peso = modelo.predict(estatura)

# Imprimir los coeficientes del modelo
print("Coeficiente de la pendiente (peso/estatura):", modelo.coef_[0])
print("Término independiente:", modelo.intercept_)

# Métricas de evaluación
mse = mean_squared_error(peso, predicciones_peso)
r2 = r2_score(peso, predicciones_peso)
mae = mean_absolute_error(peso, predicciones_peso)

print("\nMétricas de Evaluación:")
print(f"Error Cuadrático Medio (MSE): {mse:.2f}")
print(f"Coeficiente de Determinación (R²): {r2:.2f}")
print(f"Error Absoluto Medio (MAE): {mae:.2f}")

# Visualizar los resultados y el diagrama de residuos
plt.scatter(estatura, peso, color='green', label='Datos reales')
plt.plot(estatura, predicciones_peso, color='purple', linewidth=2, label='Regresión lineal')
plt.xlabel('Estatura')
plt.ylabel('Peso')
plt.title('Regresión Lineal: Estatura vs Peso')
plt.legend()
plt.show()

# Diagrama de Residuos
residuos = peso - predicciones_peso
plt.scatter(estatura, residuos)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Estatura')
plt.ylabel('Residuos')
plt.title('Diagrama de Residuos')
plt.show()
```

## Interpretación de Resultados

1. **Coeficiente de la Pendiente (peso/estatura):** El coeficiente de la pendiente es 108.99, lo que significa que, en promedio, por cada unidad adicional de estatura (en metros), se espera un aumento de aproximadamente 108.99 unidades en el peso (en la misma escala utilizada para el peso).

2. **Término Independiente:** El término independiente es -113.82, lo que indica que cuando la estatura es cero (lo cual no tiene un significado práctico en este contexto), se espera que el peso sea aproximadamente -113.82. En el contexto de este modelo, la interpretación práctica es menos relevante para el término independiente.

3. **Métricas de Evaluación:**
   - **Error Cuadrático Medio (MSE):** 9.15: Esto representa el promedio de los errores al cuadrado entre las predicciones y los valores reales. Un MSE más bajo indica que el modelo tiene buenas capacidades predictivas.
   
   - **Coeficiente de Determinación (R²):** 0.93: El 93% de la variabilidad en el peso se explica por la estatura según el modelo. Un R² cercano a 1 sugiere una fuerte relación entre las variables.
   
   - **Error Absoluto Medio (MAE):** 2.06: Esto representa el promedio de las diferencias absolutas entre las predicciones y los valores reales. Un MAE más bajo indica que el modelo hace predicciones más precisas en términos absolutos.

4. **Diagrama de Residuos**:

**Homocedasticidad**: La presencia de residuos dispersos alrededor de la línea horizontal cero indica homocedasticidad. Esto es positivo, ya que sugiere que la variabilidad de los errores es constante en diferentes niveles de estatura.

**Puntos Atípicos (Outliers)**: La presencia de residuos que se desvían significativamente de cero, especialmente aquellos arriba o por debajo del 2 en Y, podría indicar la existencia de puntos atípicos. Identificar y entender estos puntos podría ser importante para evaluar la influencia de valores extremos en el modelo.

**Patrones No Lineales**: Los residuos que varían alrededor de diferentes niveles de estatura podría indicar posiblemente patrones no lineales o la necesidad de ajustes adicionales en el modelo.

Considerando estos puntos, sería recomendable analizar más detalladamente los casos específicos asociados a los puntos atípicos o patrones no lineales. Esto podría implicar explorar datos adicionales o considerar modelos más complejos.

##Conclusiones

El análisis de regresión lineal aplicado a los datos de estatura y peso ha arrojado resultados significativos. El modelo revela una relación sólida y positiva entre ambas variables, indicando que, en promedio, el peso tiende a aumentar conforme lo hace la estatura. Las métricas de evaluación respaldan la eficacia del modelo, con un bajo Error Cuadrático Medio (MSE) y un alto Coeficiente de Determinación (R²), indicando buenas capacidades predictivas.

No obstante, es crucial abordar estas conclusiones con precaución al inferir causalidad, ya que la correlación observada no implica necesariamente una relación causal directa. Otros factores podrían influir en la dinámica entre estatura y peso.
